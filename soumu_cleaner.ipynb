```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/conversion_data/soumu_H30.csv', usecols = lambda x : x not in ['code'])

for i in range(len(df)):
    df.iloc[i] = df.iloc[i].str.replace(" ", "")
    df.iloc[i] = df.iloc[i].str.replace(",", "")
df = df.replace("①", "1")
df = df.replace("②", "2")
df = df.replace("○", "1")
df = df.fillna(0)

df.head()
```

```python
df.to_csv('data/cleaned_data/cleaned_soumu_H30.csv')
```
