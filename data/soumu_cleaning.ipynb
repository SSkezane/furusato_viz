```python
import pandas as pd
soumu = pd.read_csv("soumu.csv")
```

```python
categorical = [
    "広報活動","呼びかけ","見直し","イベントPR","決済",
    "用途選択肢","分野","事業","クラウドファンディング","まちづくり・市民活動",
    "スポーツ・文化振興","健康・医療・福祉","環境・衛生","教育・人づくり",
    "子ども・子育て","地域・産業振興","観光・交流・定住促進","安心・安全・防災","災害支援・復興"
]
# nan -> 0
soumu[categorical] = soumu[categorical].fillna(0)
# nullを探す
soumu.isnull().count()
```

```python
# cityがnullな行を削除
# 47都道府県あるのでnullの個数は1788-47 = 1741になる
soumu = soumu.dropna(axis=0,how="any")
soumu.isnull().count()
```

```python
# ○を1に置き換える
soumu = soumu.replace("○",1)
# ーを0に置き換える
soumu = soumu.replace("－",0)
# ①,②を置き換える
soumu = soumu.replace("①",1)
soumu = soumu.replace("②",2)
soumu.head()
```

```python
# nullが含まれているか
print(soumu.isnull().values.sum() != 0)
# -> False(null値の取り除き完了!)
```

```python
import numpy as np
# 型変換
integers = [
    "寄附件数","寄附金額","寄附件数_県外","寄附金額_県外",
    "返礼品等の調達に係る費用","返礼品等の送付に係る費用","広報に係る費用","決済等に係る費用","事務に係る費用","その他費用"
]
# 空白と点を削除する
for i in range(len(integers)):
    soumu[integers[i]] = soumu[integers[i]].str.replace(" ","")
    soumu[integers[i]] = soumu[integers[i]].str.replace(",","")
    soumu[integers[i]] = soumu[integers[i]].fillna(0)
soumu.to_csv("cleaned_soumu.csv",index=False)
```
