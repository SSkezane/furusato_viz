```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('data/cleaned_data/cleaned_soumu_H30.csv')
df.head()
```

```python
#Unnamedとカテゴリー変数のカラムを除外
df2 = df.iloc[0:, 1:13]
print(df2.columns)

#都道府県ごとに市町村のデータを合算 -> 都道府県の順序がバラバラになる
df_city = df2.query('city != 0')
df_sum = df_city.groupby('pre').sum()

#重複を除外し、元の順序の都道府県名リストを作成
df_pre = df2[~df2['pre'].duplicated()]
list_pre = df_pre['pre'].values.tolist()

#都道府県の順序を元に戻す
df3 = df_sum.reindex(index=list_pre)
df3.head()
```

```python
df3.to_csv('data/cleaned_data/cleaned_soumu2_H30.csv')
```
